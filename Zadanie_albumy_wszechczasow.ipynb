Import modułu "pandas" + import danych ze strony internetowej:

```python
import pandas as pd

#Wczytuję bazę ze strony internetowej / html - bazę w formacie DataFrame przypisuję to zmiennej -> table.
data = pd.read_html('https://www.officialcharts.com/chart-news/the-best-selling-albums-of-all-time-on-the-official-uk-chart__15551', header=0)
table = data[0]
table
```

1. Zmieniam nazwy nagłówków kolumn:

```python
table.rename(columns={'POS':'POZYCJA','TITLE':'TYTUŁ', 'ARTIST':'ARTYSTA', 'YEAR':'ROK', 'HIGH POSN': 'MAX POZ'},inplace=True)
table
```

2. Ilu pojedynczych artystów znajduje się na liście?

```python
print(f"Pojedynczych artystów znajduje się na liście jest {table['ARTYSTA'].nunique()}.")
```

3. Które zespoły pojawiają się najczęściej na liście? *ogarniczyłem się do 5.

```python
most_popular = table['ARTYSTA'].value_counts()[:5]
for artysta, ilosc in most_popular.items():
    print(f"'{artysta}' - ilość albumów: {ilosc}")
```

4. Zmień nagłówki kolumn, tak aby każdy z nich rozpoczynał się od wielkiej litery, a pozostałe były wprowadzone małymi literami.

```python
table.columns = table.columns.str.title()
table
```

5. Wyrzuć z tabeli kolumnę ‘Max Poz’

```python
table.drop('Max Poz', axis=1, inplace=True)
```

6. W którym roku wyszło najwięcej albumów znajdujących się na liście?

```python
table_year = table['Rok'].value_counts()
max_amount = table_year.max()
the_most_popolar_year = table_year[table_year == max_amount]
print(f"Rok lub lata w których wychodziło najwięcej albumów z listy:")
for year, amount in the_most_popolar_year.items():
    print(f"'Rok: {year}' - ilość albumów: {amount}")
```

7. Ile albumów wydanych między 1960 a 1990 rokiem włącznie znajduje się na liście?

```python
table['Rok'] = table['Rok'].astype(int)
year_1960_1990 = table[(table['Rok']>= 1960) & (table['Rok'] <= 1990)]
print(f"Między 1960 a 1990 rokiem włącznie wydano {len(year_1960_1990)} albumy.")
```

8. W którym roku wydany został najmłodszy album na liście?

```python
year_max = table['Rok'].max()
Newest_album = table[table['Rok'] == year_max]
print(f"Najmłodszy album na liście to album: '{Newest_album['Tytuł'].iloc[0]}'/ {Newest_album['Artysta'].iloc[0]} z roku: {Newest_album['Rok'].iloc[0]}")
```

9. Przygotuj listę najwcześniej wydanych albumów każdego artysty, który znalazł się na liście.

```python
table.set_index(['Artysta','Tytuł'], inplace=True)
table_sorted = table.sort_values(by=['Artysta', 'Rok'], ascending=[True, True])
print(f"Listę najwcześniej wydanych albumów każdego artysty:")
print(table_sorted[['Rok']])
```

10. Listę zapisz do pliku csv.

```python
table_sorted.to_csv('Lista_najwczesniej_wydanych_albumow_wg_artysty.csv')
```
