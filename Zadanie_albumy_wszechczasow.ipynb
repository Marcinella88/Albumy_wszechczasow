Import modu≈Çu "pandas" + import danych ze strony internetowej:

```python
import pandas as pd

#Wczytujƒô bazƒô ze strony internetowej / html - bazƒô w formacie DataFrame przypisujƒô to zmiennej -> table.
data = pd.read_html('https://www.officialcharts.com/chart-news/the-best-selling-albums-of-all-time-on-the-official-uk-chart__15551', header=0)
table = data[0]
table
```

1. Zmieniam nazwy nag≈Ç√≥wk√≥w kolumn:

```python
table.rename(columns={'POS':'POZYCJA','TITLE':'TYTU≈Å', 'ARTIST':'ARTYSTA', 'YEAR':'ROK', 'HIGH POSN': 'MAX POZ'},inplace=True)
table
```

2. Ilu pojedynczych artyst√≥w znajduje siƒô na li≈õcie?

```python
print(f"Pojedynczych artyst√≥w znajduje siƒô na li≈õcie jest {table['ARTYSTA'].nunique()}.")
```

3. Kt√≥re zespo≈Çy pojawiajƒÖ siƒô najczƒô≈õciej na li≈õcie? *ogarniczy≈Çem siƒô do 5.

```python
most_popular = table['ARTYSTA'].value_counts()[:5]
for artysta, ilosc in most_popular.items():
    print(f"'{artysta}' - ilo≈õƒá album√≥w: {ilosc}")
```

4. Zmie≈Ñ nag≈Ç√≥wki kolumn, tak aby ka≈ºdy z nich rozpoczyna≈Ç siƒô od wielkiej litery, a pozosta≈Çe by≈Çy wprowadzone ma≈Çymi literami.

```python
table.columns = table.columns.str.title()
table
```

5. Wyrzuƒá z tabeli kolumnƒô ‚ÄòMax Poz‚Äô

```python
table.drop('Max Poz', axis=1, inplace=True)
```

6. W kt√≥rym roku wysz≈Ço najwiƒôcej album√≥w znajdujƒÖcych siƒô na li≈õcie?

```python
table_year = table['Rok'].value_counts()
max_amount = table_year.max()
the_most_popolar_year = table_year[table_year == max_amount]
print(f"Rok lub lata w kt√≥rych wychodzi≈Ço najwiƒôcej album√≥w z listy:")
for year, amount in the_most_popolar_year.items():
    print(f"'Rok: {year}' - ilo≈õƒá album√≥w: {amount}")
```

7. Ile album√≥w wydanych miƒôdzy 1960 a 1990 rokiem w≈ÇƒÖcznie znajduje siƒô na li≈õcie?

```python
table['Rok'] = table['Rok'].astype(int)
year_1960_1990 = table[(table['Rok']>= 1960) & (table['Rok'] <= 1990)]
print(f"Miƒôdzy 1960 a 1990 rokiem w≈ÇƒÖcznie wydano {len(year_1960_1990)} albumy.")
```

8. W kt√≥rym roku wydany zosta≈Ç najm≈Çodszy album na li≈õcie?

```python
year_max = table['Rok'].max()
Newest_album = table[table['Rok'] == year_max]
print(f"Najm≈Çodszy album na li≈õcie to album: '{Newest_album['Tytu≈Ç'].iloc[0]}'/ {Newest_album['Artysta'].iloc[0]} z roku: {Newest_album['Rok'].iloc[0]}")
```

9. Przygotuj listƒô najwcze≈õniej wydanych album√≥w ka≈ºdego artysty, kt√≥ry znalaz≈Ç siƒô na li≈õcie.

```python
table.set_index(['Artysta','Tytu≈Ç'], inplace=True)
table_sorted = table.sort_values(by=['Artysta', 'Rok'], ascending=[True, True])
print(f"Listƒô najwcze≈õniej wydanych album√≥w ka≈ºdego artysty:")
print(table_sorted[['Rok']])
```

10. Listƒô zapisz do pliku csv.

```python
table_sorted.to_csv('Lista_najwczesniej_wydanych_albumow_wg_artysty.csv')
```
